```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
img=cv2.imread("rengoku.jpg")
#convert from BGR format to RGB format
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
X=img.reshape(-1,3)
X.shape
```

```python
#each pixel is of the form (R,G,B) value
#it means each pixel can be plotted in 3D space
k=7
model=KMeans(n_clusters=k)
model.fit(X)
```

```python
centroids=model.cluster_centers_
centroids
```

```python
colors=np.array(centroids,dtype="uint8")
colors
```

```python
i=1
for color in colors:
  plt.subplot(1,k,i)
  plt.axis("off")
  i+=1
  plt.imshow([[[color[0],color[1],color[2]]]])
plt.show()
```
